```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Hospital ER.csv')
```

```python
df = df.dropna()
```

```python
print(df.dtypes)
```

```python
# df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y %H:%M')
df['patient_id'] = df['patient_id'].astype(str)
df['patient_sat_score'] = pd.to_numeric(df['patient_sat_score'], errors='coerce')
```

```python
df.head()
```

```python
# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')

# Now you can extract the date and time components
df['new_date'] = df['date'].dt.strftime('%d %B %Y')
df['time'] = df['date'].dt.strftime('%H:%M:%S')

# Drop the original 'date' column
df = df.drop('date', axis=1)

# Rename the 'new_date' column to 'date'
df.rename(columns={"new_date": "date"}, inplace=True)
```

```python
df.head()
```

```python
df['full_name'] = df['patient_first_inital'] + ' ' + df['patient_last_name']
```

```python
df.head()
```

```python
df = df.drop('patient_first_inital', axis=1)
df = df.drop('patient_last_name', axis=1)
```

```python
df.head()
```

```python
def categorize_age(age):
    if age >= 70:
        return '70+'
    elif age <= 10:
        return '0-10'
    else:
        start = ((age - 1) // 10) * 10 + 1
        end = start + 9
        return f"{start}-{end}"
```

```python
df['age_bucket'] = df['patient_age'].apply(categorize_age)
```

```python
df.head()
```

```python
def age_group(age):
    if age <= 2:
        return 'infant'
    elif age <= 6:
        return 'early childhood'
    elif age <= 12:
        return 'middle childhood'
    elif age <= 18:
        return 'teenage'
    else:
        return 'adult'
```

```python
df['age_group'] = df['patient_age'].apply(age_group)
```
